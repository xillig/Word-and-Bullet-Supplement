```python
# Stationarity: The Augmented Dickey-Fuller (ADF) test is a common way to test for stationarity.
# The null hypothesis is that the series is non-stationary. 
# If the p-value is less than your chosen significance level (often 0.05), you can reject the null and assume the series is stationary.

# Stationarity: Non-stationarity is a common issue in time series data, where the properties of your data change over time.
# The usual approach to handling non-stationarity is to transform your data in a way that makes it stationary. Here are a few methods:
# Differencing: This is the most common method. It involves computing the difference between consecutive observations. 
# In Pandas, this can be done with the diff() function.

from statsmodels.tsa.stattools import adfuller

def adfuller_test(series):
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    return
```

```python
# No perfect multicollinearity: This can be tested by checking the correlation between different lags of your time series. 
# If any have a correlation of (-)1 or very close to (-)1, one may have a problem.

import numpy as np
import plotly.figure_factory as ff

# Concatenate your dataframes and calculate the correlation matrix
correlation_matrix = np.corrcoef(pd.concat([casulties_df[col_li], monthly_df[['1','2','4','5']]], axis=1), rowvar=False)

# Create a list of column names
columns = col_li + ['topic_' + str(i) for i in ['1', '2', '4', '5']]

# Create the annotated heatmap with Plotly
fig = ff.create_annotated_heatmap(
    z=correlation_matrix, 
    x=columns, 
    y=columns, 
    annotation_text=np.around(correlation_matrix, decimals=2), 
    showscale=True
)
fig.show()
```

```python
from statsmodels.tsa.api import VAR

import warnings
warnings.filterwarnings('ignore')

def get_aic_bic(topic_no):
    for i in col_li:
        print(str(i))
        for j in range(2,13):
            
            model = VAR(pd.concat([casulties_df[i].diff().dropna(), monthly_df[str(topic_no)].diff().dropna()], axis=1))
            res_aic = model.fit(maxlags=j, ic='aic')  # 'aic' could be replaced with 'bic' for BIC
            res_bic = model.fit(maxlags=j, ic='bic')  # 'aic' could be replaced with 'bic' for BIC
            print('AIC lag' + str(j) + ': ' + str(res_aic.aic))
            print('BIC lag' + str(j) + ': ' + str(res_bic.bic))
            print('\n')
    return

# No mayor differences between the AIC and BIC in any variable
```

```python
# null hypothesis of homoskedasticity
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.regression.linear_model import OLS
from statsmodels.tools.tools import add_constant

def test_homoscedasticity(df):
    # create an autoregressive model
    model = OLS(df.iloc[:,0], add_constant(df.iloc[:,1]))
    result = model.fit()

    # calculate residuals
    residuals = result.resid

    # perform the Breusch-Pagan test
    bp_test = het_breuschpagan(residuals, add_constant(df.iloc[:,1]))
    labels = ['LM Statistic', 'LM-Test p-value', 'F-Statistic', 'F-Test p-value']
    test_results = dict(zip(labels, bp_test))

    return test_results
```

```python
# Topics to Outgroups - heteroscedasticity for: Topic: 4, sumFar
for topic in range(1,6):
    for i in col_li:
        print("Topic: " + str(topic) + ", " + i)
        #print(test_homoscedasticity(pd.concat([pd.Series(yeojohnson(casulties_df[i])[0]).diff().dropna(), pd.Series(yeojohnson(monthly_df[str(topic)])[0]).diff().dropna()], axis=1)))
        print(test_homoscedasticity(pd.concat([casulties_df[i].diff().dropna(), monthly_df[str(topic)].diff().dropna()], axis=1)))
    print('\n')
```

```python
# Outgroups to Topics - heteroscedasticity for: Topic: 1, sumRawafid; Topic: 4, sumRawafid; Topic: 4, sumNear2
for topic in range(1,6):
    for i in col_li:
        print("Topic: " + str(topic) + ", " + i)
        print(test_homoscedasticity(pd.concat([monthly_df[str(topic)].diff().dropna(), casulties_df[i].diff().dropna()], axis=1)))
    print('\n')
```

```python
# The Shapiro-Wilk test: null hypothesis is that the sample has been generated from a normal distribution

import numpy as np
import pandas as pd
import statsmodels.tsa.api as smt
from scipy.stats import shapiro

def check_normality(df):
    model = smt.VAR(df)
        
    results = model.fit(maxlags=6, ic='aic')
    residuals = results.resid

    # Shapiro-Wilk test for each series of residuals
    test_results = {name: shapiro(resid) for name, resid in residuals.items()}

    return test_results
```

```python
# Topics to Casulties:
# Normality of the residuals cannot be assumed for all variables. Use F-Test statistics of Granger Causality Test with caution, use Chi-Squared-based test statistics instead!
for topic in range(1,6):
    for i in col_li:
        print("Topic: " + str(topic) + ", " + i)
        print(check_normality(pd.concat([casulties_df[i].diff().dropna(), monthly_df[str(topic)].diff().dropna()], axis=1)))
    print('\n')
```

```python
# Casulties to Topics:
# Normality of the residuals cannot be assumed for all variables. Use F-Test statistics of Granger Causality Test with caution, use Chi-Squared-based test statistics instead!
for topic in range(1,6):
    for i in col_li:
        print("Topic: " + str(topic) + ", " + i)
        print(check_normality(pd.concat([monthly_df[str(topic)].diff().dropna(), casulties_df[i].diff().dropna()], axis=1)))
    print('\n')
```
